**Технологический процесс:**
- Rougher feed — исходное сырье

- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота


**Параметры этапов:**
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

### Проверка эффективности обогащения

Проверим верность рассчета эффективности обогащения. Используем для проверки обучающий набор по признаку rougher.output.recovery.

```python
def gold_recovery(data):
    C = data['rougher.output.concentrate_au']
    F = data['rougher.input.feed_au']
    T = data['rougher.output.tail_au']
    
    recovery = (C * (F - T)) / (F * (C - T)) * 100
    return recovery
```

 Найдём MAE между нашими расчётами и значением признака.

```python
mean_absolute_error(train['rougher.output.recovery'], gold_recovery(train))
```

## Пропуски данных и нулевые значения

В качестве гипотезы, примем положение, что нулевые суммы явлюятся ошибочными данными, которые могут повлиять на результаты. Создадим дублирующий фрейм, в котором будут удалены эти показатели, чтобы провести обучение алгоритмов на двух вариантах данных.

```python
def cleaner_slave(data, filter):
    '''
    Функция cleaner_slave принимает датасет (data),
    и фильтр (filter), состоящий из обрабатываемых колонок датасета (data).
    После чего, оценивает каждую колонку на предмет выбросов, соответствующих 1.5 IQR
    и очищает датасет (data) от этих данных.
    '''
    # выведем минимальные и максимальные значения
    for inx, col in enumerate(filter):
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_fence = Q1 - (1.5 * IQR)
        upper_fence = Q3 + (1.5 * IQR)
        
        # выведем статистику
        print('\033[1m' + 
              f'{inx}. Пороговые значения оцениваемого признака {filter[inx]}:' 
              + '\033[0m')
        print('Нижняя граница значений:', lower_fence)
        print('Верхняя граница значений:', upper_fence)
        print('Количество значений за пределом нижней границы:', len(data[data[col] < lower_fence]))
        print('Количество значений за пределом верхней границы:', len(data[data[col] > upper_fence]))
        print()
        
        # произведем очистку датасета              
        data = data.loc[(data[col] >= lower_fence) & (data[col] <= upper_fence)]
     
    return data
```

## Модель

Подготовим функцию для оценки модели. Для решения задачи введём новую метрику качества — sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).
Она похожа на MAE, но выражается не в абсолютных величинах, а в относительных.

```python
def smape(y_true, y_pred):
    y_true = np.ravel(y_true)
    y_pred = np.ravel(y_pred)
    total = 0
    for i in range(len(y_true)):
        total += abs(y_true[i] - y_pred[i]) / ((abs(y_true[i]) + abs(y_pred[i]))/2)
        smape = abs(total*100 / len(y_true))
    return abs(smape)
```

```python
def smape_final(smape_rougher, smape_final):
    smape_final = (0.25 * smape_rougher + 0.75 * smape_final)
    return smape_final
```
