# Отток клиентов фитнес-центра

## Описание задачи

### Постановка задачи

Заказчиком выступает сеть фитнес-центров «Культурист-датасаентист». Результаты обучения алгоритмов МО будут использованы при разработке стратегии взаимодействия с клиентами на основе аналитических данных.


Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. Конечно, не исключено, что он уехал на Бали и по приезде обязательно продолжит ходить на фитнес. Однако чаще бывает наоборот. Если клиент начал новую жизнь с понедельника, немного походил в спортзал, а потом пропал — скорее всего, он не вернётся.
Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет. 

**Задача** 
Провести анализ и подготовить план действий по удержанию клиентов:
1. Научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
2. Сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
3. Проанализировать основные признаки, наиболее сильно влияющие на отток;
4. сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:
        4.1. Выделить целевые группы клиентов;
        4.2. Предложить меры по снижению оттока;
        4.3. Определить другие особенности взаимодействия с клиентами.

### Ход исполнения

1. Загрузить и подготовить данные.
2. Провести исследовательский анализ данных (EDA).
3. Построить модель прогнозирования оттока клиентов.
4. Провести кластеризацию клиентов.
5. Сформулировать выводы и сделать базовые рекомендации по работе с клиентами.

### Сводный анализ

Сравним средние значения признаков по трем массивам.

```python
# подготовим список
churn_list = [df, df_away, df_stay]
churn_name = ['общем', 'ушедших', 'оставшихся']
```

```python
def subtraction_slave(sub, base = df):
    print('\033[1m' + f'Разница средних значений признаков' + '\033[0m')
    display(base.describe().loc[['mean']] - sub.describe().loc[['mean']]  )
```

Оценим изменения средних значений и корреляции между всеми массивами.

Начнем с различий между общим и ушедшими.

```python
subtraction_slave(churn_list[1])
```

**Выводы:** в целом, почти все признаки имеют большее среднее значение, что укладывается в логику.

Различия между общим и оставшимися.

```python
subtraction_slave(churn_list[2])
```

**Выводы:** все признаки для оставшихся имеют более низкое среднее значение, чем для общего массива. Это может означать, что ушедшие склонны к высоким значениям. Возможно наличие выбросов.

Различия между ушедшими и оставшимися.

```python
subtraction_slave(churn_list[1], churn_list[2])
```

**Выводы:** можно с уверенностью предположить, что ушедшие более склонны к демонстрации крайних максимальных значений в описании признаков.

```python
def boxplot_slave(data, name, xlim = None, ylim = None):
    '''
    boxplot_slave принимает данные в параметр data и возвращает боксплот;
    name используется для разметки и принимает название исследуемого признака.
    xlim и ylim используются для масштабирования графика, ограничивая его оси
    и принимают кортеж (tuple) из двух цифр (начало и конец ограничения)
    '''
    # выведем описание данных
    print('\033[1m' + f'Описание признака {name}:' + '\033[0m')
    print(data.describe())
    print('=' * 45)
    
    # построим боксплот
    box = plt.boxplot(
        x=data, vert=False, 
        patch_artist=True, notch=True
    )
    fig = plt.gcf()
    fig.set_size_inches(15, 8)
    if xlim is not None:
        plt.xlim(xlim)
    if ylim is not None:
        plt.ylim(ylim)
    plt.title(
            f'Боксплот признака {name}', 
            fontsize=14
        )
    plt.show()
    
    # выведем его минимальные и максимальные значения
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - (1.5 * IQR)
    upper_fence = Q3 + (1.5 * IQR)
    print('\033[1m' + f'Пороговые значения боксплота {name}:' + '\033[0m')
    print('Нижняя граница значений:', lower_fence)
    print('Верхняя граница значений:', upper_fence)
    print('Количество значений за пределом нижней границы:', len(data[data < lower_fence]))
    print('Количество значений за пределом верхней границы:', len(data[data > upper_fence]))
    print()
    print()
```

```python
def value_changer(data, name, operation, number):
    '''
    Функция принимает data, после чего, в зависимости от 
    выбранного типа operation (параметр принимает строку 'median' или 'mean',
    для расчета, опираясь на, соответственно, медианное значение или среднее),
    чтобы для признака name заменить все выбросы, убегающие вправо,
    отфильтрованные по числу number.
    '''
    # произведем замену оставшихся
    if operation == 'median':
        value_0 = data[data['churn'] == 0][name].median()
        value_1 = data[data['churn'] == 1][name].median()
    elif operation == 'mean':
        value_0 = data[data['churn'] == 0][name].mean()
        value_1 = data[data['churn'] == 1][name].mean()
    else:
        print('Неправильный параметр operation. Используй строку \'median\' или \'mean\'')
    filter_0 = (data[name] > number) & (data['churn']==0)
    data.loc[filter_0, name] = value_0
    
    # произведем замену ушедших    
    filter_1 = (data[name] > number) & (data['churn']==1)
    data.loc[filter_1, name] = value_1
```

### Функция проверки ml_slave

Определим функцию проверки модели

```python
# определим функцию проверки
def ml_slave(mod, data_list, score_list, fit_model):
    '''
    Функция принимает разные модели в параметры, чтобы вывести их accuracy и сравнить с целевым.
    mod - вид модели;
    data_list - список, содержащий датасеты для обучения модели [X_train, y_train, X_test, y_test];
    score_list - список, собирающий оценки модели;
    fit_model - список, куда сохраняются обученные модели.
       
    '''
    # обучим модель
    model = mod.fit(data_list[0], data_list[1]) 
    predictions = model.predict(data_list[2])
    recall = round(recall_score(data_list[3], predictions), 2)
    precision = round(precision_score(data_list[3], predictions), 2)
    accuracy = round(accuracy_score(data_list[3], predictions), 2)
    f1 = round(f1_score(data_list[3], predictions), 2)
    fit_model = fit_model.append(model)
    
    # сохраним оценки для анализа
    score_list.append(recall)
    score_list.append(precision)
    score_list.append(accuracy)
    score_list.append(f1)
   
    # выведем метрики
    print(f'Recall модели: {recall}')
    print(f'Precision модели: {precision}')
    print(f'Accuracy модели: {accuracy}')
    print(f'f1 модели: {f1}')
    print('Лучшие параметры модели:', model.best_params_)
    print('-' * 45)
    
    # проверка на адкеватность
    # используем константное значение - медиану
    # создадим список стратегий
    strategies = ['most_frequent', 'stratified', 'uniform']

    # проверим accuracy по всем стратегиям
    for strat in strategies:
        dc = DummyClassifier(strategy=strat, random_state=42)
        dc.fit(data_list[0], data_list[1])
        predictions = dc.predict(data_list[2])
        res = round(dc.score(predictions, data_list[3]), 2)
        print('Испытываемая стратегия: ', {strat:res})
        print('Разница оценки accuracy модели и DC:', accuracy - res)
        if accuracy > res:
            print('Вывод: обученная модель лучше')
        else:
            print('Вывод: обученная модель хуже')
         
    # техническая разметка
    print('=' * 45)
    print()    
```

```python
def feature_importanter(data, model_name):
    sns.set(rc={'figure.figsize':(11.7,8.27)})
    feature_importance = data
    feature_imp = pd.DataFrame(sorted(zip(feature_importance, features)), columns=['Value','Feature'])
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False))
    plt.title(f'Feature importance для {model_name}', fontsize=16)
    plt.tight_layout()
    plt.show();
```

```python
feature_importanter(fit_model[1].best_estimator_.named_steps['rfc'].feature_importances_, models[1])
```

Косвенным подтверждением этого факта является ценность признаков. Мы видим, что RFC при принятии решения опирается на наш синтетичесий признак, что в целом логично. Чем чаще клиент посещает зал, тем меньше вероятность его отказа. Вторым важным признаком для этой модели является общая длительность посещения зала, что вполне коррелирует с первым признаком и формирует интерпретируемую картину прогноза. Обратим внимание, что указанные признаки имеют наибольшую корреляцию с целевым.

При этом LR модель увеличивает веса наличия партнера, близость локации и приведенного друга. Эти факторы имеют низкую корреляцию с целевым. Однако, могут быть использованы как вторичные для прогнозирования и выстраивания коммерческой стратегии.

Таким образом, учитывая, что глубокая детализация скорее повредит, сформируем 5 кластеров пользователей, основываясь на значении К приблизительно 55.

### K-means

#### Формирование кластеров

```python
# km_pipe
km_pipe = Pipeline([
    # стандартизация признаков
    ('std_caler', StandardScaler()),
    
    # работа с моделью, определим количество кластеров 5
    ('km', KMeans(n_clusters=5, random_state=42))
])
```

```python
km_pipe
```

```python
# создадим новый датасет, содержащий столбец с номером кластера
df_clusters = df.copy()
df_clusters['cluster'] = km_pipe.fit(X).named_steps['km'].labels_
```

#### Средние значения признаков кластеров

Изучим средние значения образованных кластеров.

```python
display(df_clusters.groupby('cluster').agg('mean'))
df_clusters['cluster'].value_counts()
```

```python
# процент оттока по кластерам
df_clusters.groupby('cluster').agg('mean')['churn'].sort_values(ascending=False)
```

- В 1 кластере собраны люде, находящиеся удаленно от фитнес-центра. Они занимают второе место по оттоку клиентов.
- Во 2 кластере, люди, живущие поблизости и посещяющие групповые занятия. Они занимают второе место по лояльности. 
- В 3 кластер имеет самое низкое значение оттока. Самый многочисленный кластер. Для обеспечения лояльности клиентов, стоит изучить его призанки.
- 4 кластер никогда не посещает групповые занятия и занимает третье место по оттоку клиентов.

Признаки 0 и 1 кластеров могут помочь спрогнозировать отток клиентов.

Вторая группа по оттоку. Все они расположены удаленно от фитнес-центра, что несомненно является критическим фактором. Их активность чуть выше 0 кластера, но при этом, временные метрики одни из самых плохих. Дополнительные траты занимают предпоследнее место.

**КЛАСТЕР 2**

Все клиенты из этого кластера имеют фитнес-центр рядом, а также высокую временную активность. Их контракты и лайфтайм (имеет длинный хвост убывающих значений вправо) стремсятся к более высокой длительности. Кроме того, они имеют 100% посещаемость групповых занятий. Кластер, имеющий отток всего в 6%! Почти все приведены по промо-коду или партнером клуба. Самая большая длительность контракта и частота посещений (уступает только кластеру 2).
