## Анализ данных

Создадим инструменты для анализа. После анализа объединим данные в один датасет и проведем предобработку.

```python
class DateAnalyst():
    '''
    Класс служит для анализа и обзора данных, 
    и использует следующие методы:
    informer для вывода info() и head() по заданному сету;
    null_finder для поиска пропусков в колонке;
    values для подсчета значений в колонке;
    uniq для проверки данных в колонке на уникальность.
    '''
    def __init__(self, df, name):
        self.df = df
        self.name = name
        pass
    
    # обзор данных    
    def informer(self):
        print('\033[1m' + f'Исследуем датасет {self.name}' + '\033[0m')
        display(self.df.head(10))
        self.df.info()
    
    # поиск пропусков в колонках датасета   
    def null_finder(self):
        print('\033[1m' + f'Исследуем датасет {self.name}' + '\033[0m')
        null_list = []
        for inx, column in enumerate(self.df):
            null_count = self.df[self.df.columns[inx]].isna().sum()
            print(f'Исследуем колонку {self.df.columns[inx]}')
            print(f'Количество пропусков: {null_count}')
            if null_count > 0:
                null_list.append(self.df.columns[inx])
                print('=' * 45)
                print()
        if len(null_list) > 0:
            print('\033[1m' + 'Колонки с пропусками:' + '\033[0m', null_list)
    
    # проверка значений в колонке на типовость
    def values(self, columns):
        print('\033[1m' + f'Исследуем датасет {self.name}' + '\033[0m')
        for column in columns:
            print(f'Значения в колонке{column}')
            print(self.df[column].value_counts())
            print('=' * 45)
            print()
    
    # проверка значений в колонке на уникальность
    def uniq(self, columns: list):
        print('\033[1m' + f'Исследуем датасет {self.name}' + '\033[0m')
        for column in columns:
            print(f'Количество уникальных значений в колонке {column}: ', 
                 self.df[column].nunique())
            print(f'При общем количестве значений: ',
                 self.df[column].count())
            print(f'Итого, количество неуникальных значений: ',
                 self.df[column].count() - self.df[column].nunique())
            print('=' * 45)
            print()
```

```python
class CatDescriber():
    def __init__(self, df=df):
        self.df = df
        pass
    
    def scoreboard(self, column):
        self.column = column
        # визуализаци распределения
        plt.figure(figsize=(12,6))
        self.df[self.column].hist(ec='black', alpha=0.84)
        self.df[self.df['label'] == 1][self.column].hist(ec='black', alpha=0.84)
        plt.legend(['Все пользователи', 'Отток'])
        plt.xticks(rotation=42)
        plt.title(f'Отток по признаку {self.column}')
        plt.ylabel('Количество человек')
        plt.show()
        
        # таблица соотношения категории признака и метки оттока "1"
        ratio = (self.df.groupby(column)['label'].agg(['sum', 'count'])
         .assign(ratio = lambda x: x['sum'] / x['count'])
        ).sort_values(by='ratio', ascending=False)
        
        return ratio.style.background_gradient(low=0.42).set_caption(
            f'Таблица соотношения категорий признака {self.column} и метки оттока 1'
        )
```

```python
# вернем данные, сбросив индекс и преобразуем названия колонок в единый тип
def col_name(data):
    data.reset_index(inplace=True)
    data.columns = ['customer_id', 'begin_date', 'end_date', 'type', 'paperless_billing', 'payment_method',
       'monthly_charges', 'total_charges', 'internet_service', 'online_security',
       'online_backup', 'device_protection', 'tech_support', 'streaming_tv',
       'streaming_movies', 'gender', 'senior_citizen', 'partner', 'dependents',
       'multiple_lines']
    return data
```

```python
# заполним пропуски
def fillnan(data):
    cols = ['online_security', 'tech_support', 'internet_service', 'online_backup', 
            'device_protection', 'streaming_movies', 'streaming_tv', 'multiple_lines']
    data[cols] = data[cols].fillna('No')
    
        
    return data
```

```python
# выделим целевой признак и изменим тип данных
def change_types_and_label(data):
    data.loc[data['total_charges'] == ' ', 'total_charges'] = data['monthly_charges']
    data['total_charges'] = data['total_charges'].astype('float64')
    data[data.columns.drop(
    ['begin_date', 'end_date', 'total_charges', 'monthly_charges']
    )] = data[data.columns.drop(
    ['begin_date', 'end_date', 'total_charges', 'monthly_charges']
    )].astype('category')
    data['label'] = np.where(data['end_date'] == 'No', 0, 1).astype('int64')
        
    return data
```

```python
# дополним новыми признаками
def chaneg_new_features(data):
    data['end_date_02_20'] = data['end_date'].apply(lambda x: '2020-02-01 00:00:00' if x =='No' else x)
    data['end_date_02_20'] = pd.to_datetime(data['end_date_02_20'], format='%Y-%m-%d %S:%M:%H')
    data['total_days'] = (data['end_date_02_20'] - data['begin_date']).astype('timedelta64[D]')
    data['total_days'] = data['total_days'].astype('int64')
    
    data['delta_charges'] = data['total_charges'] / data['monthly_charges']
    
    return data
```

```python
# избавимся от лишних колонок
def drop_cols(data):
    cols = ['customer_id', 'begin_date', 'end_date', 'end_date_02_20']
    data.drop(cols, inplace=True, axis=1)
    
    return data
```

```python
# запустим
df_prep = (df
    .pipe(col_name)
    .pipe(fillnan)
    .pipe(change_types_and_label)
    .pipe(chaneg_new_features)
    .pipe(drop_cols)
)
```

### Подготовка датасета к МО

Разобьем выборки на train и test, чтобы избежать утечки при препроцессинге.

```python
class MakeSet():
    '''
    Класс MakeSet служит для подготовки данных к МО,
    и использует следующие методы:
    target_features_split делит набор на X и y;
    cat_num_split - выделяет категории признаков, возвращая
    cluster_features, num_features, cat_features, binary_features;
    train_test_split - делит набор на учебный и тестовый.
    '''
    def __init__(self, df, name=False):
        self.df = df
        if name:
            self.name = name
        pass
    
    # обзор данных    
    def target_features_split(self, target, features):
        target = target
        features = features
        self.X = self.df[features]
        self.y = self.df[target]
        return self.X, self.y
    
    def cat_num_split(self, num_features = ['int64', 'float64']):
        cluster_features = ['monthly_charges', 'total_charges']
        num_features = list(self.X.select_dtypes(include=num_features).columns)
        ternary_features = ['payment_method', 'type']
        binary_features = list(
            set(
                list(X_train.select_dtypes(exclude=['int64', 'float64']).columns)
            ) - set(ternary_features)
        )
        cat_features = ternary_features + binary_features
        return cluster_features, num_features, cat_features, binary_features
    
    def train_test_split(self, test_size=.2):
        X_train, X_test, y_train, y_test = train_test_split(self.X, self.y, test_size=test_size, 
                                                            stratify=self.y, random_state=42)
        return X_train, X_test, y_train, y_test
```

```python
# зададим признаки
target = 'label'
features = ['total_days', 'type', 'paperless_billing', 
            'payment_method', 'online_security', 'tech_support', 
            'senior_citizen', 'partner', 'dependents', 
            'delta_charges', 'monthly_charges', 'total_charges']
```

### Препроцессинг

Pipeline будет состоять из следующих компонентов:

1. StandardScaler для числовых признаков
2. OrdinalEncoder для порядковых признаков
3. OneHotEncoder для бинарных признаков
4. Kmean для создания дополнительного признака в виде кластера
5. delete_cols_array для удаления лишних признаков

```python
def delete_cols_array(X):
    obj = np.array([2, 3, 15, 16])
    X = np.delete(X, obj, 1)
    return X
```

```python
def return_DF(X):
    '''
    Возвращает датафрейм, вместо матрицы.
    '''
    onehot_columns = list(
        pipe.named_steps['preprocessor']
        .named_transformers_['category']
        .get_feature_names(input_features=cat_features)
    )
    
    total = num_features[:2] + onehot_columns + ['cluster']
    
    X = pd.DataFrame(X, columns=total)
    
    return X    
```

```python
class MakeCluster(BaseEstimator, TransformerMixin):
    '''
    Класс MakeCluster для добавления кластеров.
    Используя KMeans, он рассчитывает и добавляет в датасет
    3 кластера.
    '''
    
    def __init__(self, attr = True):
        self.attr = attr

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        kmeans = KMeans(n_clusters=3, random_state=42)
        X['cluster'] = kmeans.fit_predict(X)
        X['cluster'] = X['cluster'].astype("category")
        
        return X
```

```python
# полный пайплайн
preprocessor = ColumnTransformer(
    transformers = [
        ('std_scaler', StandardScaler(), num_features),
        ('category' , OneHotEncoder(drop='first'), cat_features),
        ('kmeans', MakeCluster(), cluster_features)
        ]
)
```

```python
pipe = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
        ('delete_cols_array', FunctionTransformer(delete_cols_array, validate=False)),
        ('return_DF', FunctionTransformer(return_DF, validate=False))
    ]
)
```

### Первичные модели

```python
# подготовим таблицу для оценок
models_score = {}
models_score['model_name'] = []
models_score['roc_auc'] = []
models_score['accuracy'] = []
```

```python
def ml_slave(model, name_mod=False, main_features=False):
    '''
    Функция ml_slave принимает определенную модель МО, 
    после чего производит препроцессинг, согласно подготовленному пайплайну,
    обучает модель, формирует предикты для оценки, 
    после чего обновляет таблицу оценок и выводит график roc_auc.
    '''
    # добавим имя в таблицу
    if name_mod:
        model_name = str(model.__class__.__name__)+name_mod
    else:
        model_name = str(model.__class__.__name__)
    models_score['model_name'].append(model_name)
    
    # подготовим данные
    if main_features:
        X_train_prep = pipe.fit_transform(X_train)[main_features]
        X_test_prep = pipe.transform(X_test)[main_features]
    else:
        X_train_prep = pipe.fit_transform(X_train)
        X_test_prep = pipe.transform(X_test)
    
    # обучим модель
    model = model
    model.fit(X_train_prep, y_train)
    
    # предскажем
    pred = model.predict_proba(X_test_prep)
    pred_one =pred[:,1]   
    
    # метрики
    roc_auc = roc_auc_score(y_test, pred_one)
    acc = accuracy_score(y_test, (pred_one >= 0.5))
 
    # добавим оценки в таблицу
    models_score['roc_auc'].append(roc_auc)
    models_score['accuracy'].append(acc)
    
    # вывод метрик
    print('=' * 45)
    print(f'Максимальное значение roc_auc: {roc_auc:.4f}')    
    
    print('=' * 45)
    print(f'Максимальное значение accuracy: {acc:.4f} при пороге 0.5')    
    print('=' * 45)
    
    print('Изменение точности и полноты, в зависимости от порога:')
    for threshold in np.arange(0, 0.95, 0.05):
        predicted_valid = pred_one > threshold
        precision = precision_score(y_test, predicted_valid)
        recall = recall_score(y_test, predicted_valid)
        f1 = f1_score(y_test, predicted_valid)
        print('Порог = {:.2f} | Точность = {:.3f}, Полнота = {:.3f} | F1-мера = {:.3f}'.format(
            threshold, precision, recall, f1))
    
    # вывод графика
    fpr, tpr, thresholds = roc_curve(y_test, pred_one)
    plt.figure(figsize=(10, 10))
    plt.plot(fpr, tpr, linestyle='-')
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая')

    plt.show()
```

```python
def feature_importance(model):
    feature_imp = pd.DataFrame(sorted(zip(model.feature_importances_, prep.columns)), columns=['Value','Feature'])
    
    plt.figure(figsize=(20, 10))
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False))
    plt.title(f'{model.__class__.__name__} Features', fontsize=18)
    plt.yticks(fontsize=16)
    plt.show()
```

```python
def main_features(model, threshold):
    feature_imp = pd.DataFrame(sorted(zip(model.feature_importances_, prep.columns)), columns=['Value','Feature'])
    main_features = feature_imp[feature_imp['Value'] >= threshold]['Feature']
    
    return list(main_features) 
```

**МОДЕЛЬНАЯ РАБОТА**

Для автоматизации процесса нами был подготовлен черновой pipeline загрузки сета данных, Pipeline обработки признаков - это позволит в дальнейшем автоматизировать процесс обработки новой информаци. Pipeline подготовки признаков позволяет провести кодирование категориальных переменных, стандартизировать количественные, в связи с их не нормально распределенной природой.

Было создано три синтетических признака:
1. Пропорция трат, чтобы избежать мультиколлинеарности двух зависимых признаков (как соотношение monthly_charges и total_charges).
2. Срок жизни клиента. При этом использовать даты начала и конца контракта не стали, чтобы избежать утечек.
